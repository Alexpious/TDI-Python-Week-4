```python
import pandas as pd
df = pd.read_csv (r"C:\Users\USER\Downloads\TDI python week3 dataset(Titanic Analysis).csv" )
num = df.select_dtypes(include=['float64' , 'int64']).columns
df[num] = df[num].fillna(df[num].median())

non= df.select_dtypes(exclude=['number']).columns
df.dropna(subset=non, inplace= True)

df['FamilySize'] = df['sibsp'] + df['parch']
df['AgeGroup'] = pd.cut(df['age'], bins=[0, 18, 40, 80], labels=['child', 'adult', 'senior'])
fildf = df[(df['survived'] == 1) & (df['pclass'] == 1)]
fildf
```

```python
# Group data by specific categories (e.g., passenger class) and perform aggregation functions such as calculating the average fare and the survival rate for each group.
import pandas as pd
df = pd.read_csv (r"C:\Users\USER\Downloads\TDI python week3 dataset(Titanic Analysis).csv" )
grouped = df.groupby('pclass').agg({'fare': 'mean', 'survived': 'mean'})
print(grouped)
```
